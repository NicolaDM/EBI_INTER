```python
import os
import pandas as pd
import numpy as np

import plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
from scipy.stats import gaussian_kde
from plotly.offline import iplot, plot
import plotly.offline as pyo
```

## 1.All positions

```python
Folder_col_all = '/nfs/research/goldman/zihao/Datas/p1_errorsProject_P4/Folder_combFor_allPos/allPos_combForcol.txt'
Folder_vir_all = '/nfs/research/goldman/zihao/Datas/p1_errorsProject_P4/Folder_combFor_allPos/allPos_combForvir.txt'

Folder_col_err = '/nfs/research/goldman/zihao/Datas/p1_errorsProject_P4/Folder_combFor_errPos/errPos_combForcol.txt'
Folder_vir_err = '/nfs/research/goldman/zihao/Datas/p1_errorsProject_P4/Folder_combFor_errPos/errPos_combForvir.txt'
```

```python
df_col_all = pd.read_csv(Folder_col_all, sep='\t')
print(df_col_all.head())
print('The number of data contained： ', len(df_col_all))
print('====================================================================================')
print('====================================================================================')
df_vir_all = pd.read_csv(Folder_vir_all, sep='\t')
print(df_vir_all.head())
print('The number of data contained： ', len(df_vir_all))
print('====================================================================================')
print('====================================================================================')
```

```python
# 1.Same as REF——Col
df_col_REF_all = df_col_all[df_col_all['REF'] == df_col_all['nucleotide_origin']]
# 2.Same as ALT——Col
df_col_ALT_all = df_col_all[df_col_all['ALT'] == df_col_all['nucleotide_origin']]
# 3.Not the same as either——Col
df_col_none_all = df_col_all[(df_col_all['ALT'] != df_col_all['nucleotide_origin']) & (df_col_all['REF'] != df_col_all['nucleotide_origin']) & (df_col_all['nucleotide_origin'] != 'N') & (df_col_all['nucleotide_origin'] != '-')]
# 4.masked——Col
df_col_masked_all = df_col_all[(df_col_all['ALT'] != df_col_all['nucleotide_origin']) & (df_col_all['REF'] != df_col_all['nucleotide_origin']) & ((df_col_all['nucleotide_origin'] == 'N') | (df_col_all['nucleotide_origin'] == '-'))]
```

```python
# 1.Same as REF——Vir
df_vir_REF_all = df_vir_all[df_vir_all['REF'] == df_vir_all['nucleotide_martin']]
# 2.Same as ALT——Vir
df_vir_ALT_all = df_vir_all[df_vir_all['ALT'] == df_vir_all['nucleotide_martin']]
# 3.Not the same as either——nucleotide_martin
df_vir_none_all = df_vir_all[(df_vir_all['ALT'] != df_vir_all['nucleotide_martin']) & (df_vir_all['REF'] != df_vir_all['nucleotide_martin']) & ((df_vir_all['nucleotide_martin'] == 'A') | (df_vir_all['nucleotide_martin'] == 'G') | (df_vir_all['nucleotide_martin'] == 'C') | (df_vir_all['nucleotide_martin'] == 'T'))]
# 4.masked——Col
df_vir_masked_all = df_vir_all[(df_vir_all['ALT'] != df_vir_all['nucleotide_martin']) & (df_vir_all['REF'] != df_vir_all['nucleotide_martin']) & (df_vir_all['nucleotide_martin'] != 'A') & (df_vir_all['nucleotide_martin'] != 'G') & (df_vir_all['nucleotide_martin'] != 'C') & (df_vir_all['nucleotide_martin'] != 'T')]
```

## 3. PLOT

```python
###############################################################################
## 1. Viridian's assembly & Colman's assembly 's nucleotides Same as reference's
###############################################################################
hist_data = [(df_col_REF_all['AF'] * 100).tolist(), 
             (df_vir_REF_all['AF'] * 100).tolist(),
             (df_col_REF_err['AF'] * 100).tolist(),
             (df_vir_REF_err['AF'] * 100).tolist()]

group_labels = ['allPosCOL-Same as REF', 'allPosVIR-Same as REF', 
                'errPosCOL-Same as REF', 'errPosVIR-Same as REF',] # name of the dataset
colors = ['#DAC6A3', '#BCD2C5', '#FBDE9C', '#A1CEF0']

fig = make_subplots(rows=2, cols=1, row_heights=[0.7, 0.3], shared_xaxes=True)

# 1.Add Density Plot subplot
for data, label, color in zip(hist_data, group_labels, colors):
    kde = gaussian_kde(data)
    x_vals = np.linspace(min(data), max(data), 1000)
    y_vals = kde(x_vals)
    
    fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', fill='tozeroy', showlegend=False, line_color=color), 
                  row=1, col=1)

# 2.Add Box Plot subplots
fig.add_trace(go.Box(x=(df_vir_REF_err['AF'] * 100).tolist(), orientation='h', name=group_labels[3], line_color=colors[3], fillcolor=colors[3]), row=2, col=1)
fig.add_trace(go.Box(x=(df_col_REF_err['AF'] * 100).tolist(), orientation='h', name=group_labels[2], line_color=colors[2], fillcolor=colors[2]), row=2, col=1)
fig.add_trace(go.Box(x=(df_vir_REF_all['AF'] * 100).tolist(), orientation='h', name=group_labels[1], line_color=colors[1], fillcolor=colors[1]), row=2, col=1)
fig.add_trace(go.Box(x=(df_col_REF_all['AF'] * 100).tolist(), orientation='h', name=group_labels[0], line_color=colors[0], fillcolor=colors[0]), row=2, col=1)

# 3.Setting up the graphic layout
fig.update_layout(
    title="Comparison of allele frequency density plots for samples - Same as reference",
    yaxis=dict(
        title='Density',
        showline=True, showgrid=False,
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        )
    ),
    xaxis=dict(
        title='Allele Frequency %',
        range=[0, 25],
        showline=True, showgrid=False,
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        )
    ),
    legend=dict(
        traceorder="normal",
        font=dict(size=12),
    ),
    plot_bgcolor='white',
    yaxis_gridcolor='lightgray', yaxis_gridwidth=0.5,
    xaxis_gridcolor='lightgray', xaxis_gridwidth=0.5,
)

# fig.show()
width = 1200
height = 900


"""
# Specify the filename with .html extension
html_file = 'Figure/PLOT_sameAsREF.html'

# Use plotly.offline.plot to save the figure as an interactive HTML file
pyo.plot(fig, filename=html_file, auto_open=False, include_plotlyjs='cdn', show_link=False)
"""
# Save the figure as an image file and specify the width and height
pio.write_image(fig, '/nfs/research/goldman/zihao/Datas/p1_errorsProject_P4/Folder_visualisation/PLOT_sameAsREF.png', width=width, height=height)
pio.write_image(fig, 'Figure/PLOT_sameAsREF.png', width=width, height=height)
```

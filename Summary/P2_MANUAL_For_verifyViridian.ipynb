## 2.2_Combine_into_one_DF
[Return to Catalog](#Catalog:)
- **Idea**:
    - label_masked: The values in this column depend on whether the values in the nucleotide_martin and nucleotide_origin columns are within the specified character lists. If the value in the nucleotide_martin column is among ['-', 'n', 'm', 'r', 'w', 's', 'y', 'k', 'v', 'h', 'd', 'b'], and the value in nucleotide_origin is among ['-', 'n'], then the value in this column will be 1, otherwise 0.

    - label_mar: The values in this column are determined by whether the value in the nucleotide_martin column is within the specified character list. If it's among ['-', 'n', 'm', 'r', 'w', 's', 'y', 'k', 'v', 'h', 'd', 'b'], then the value in this column will be 1, otherwise 0.

    - label_ori: The values in this column depend on whether the value in the nucleotide_origin column is in the list ['-', 'n']. If it is, then the value in this column will be 1, otherwise 0.

    - label_same: The values in this column depend on whether the values in the nucleotide_martin and nucleotide_origin columns are equal and whether they're within the specified character list. If they're equal and are in the list ['-', 'n', 'm', 'r', 'w', 's', 'y', 'k', 'v', 'h', 'd', 'b'], then the value in label_same will be 2. If they're equal but not in the specified list, then the value in label_same will be 1. If they're not equal, then it checks if the value in label_masked, label_mar, or label_ori is 1. If yes, then label_same will be 2, otherwise, it will be 0.

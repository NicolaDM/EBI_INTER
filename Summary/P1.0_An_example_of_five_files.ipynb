```python
import pandas as pd
import gzip
import os
import urllib.request
import pysam
```

```python
# 读取文件
df = pd.read_csv('../Datas/File_5/File_5_whole.txt', delimiter='\t', nrows=10)
```

```python
pd.set_option('display.max_colwidth', None)
df
```

```python
df_sample = df[df['run_id'] == 'SRR21269155']
df_sample
```

```python
# Download exemple
# Specify the folder path
folder_path = '../Folder/File_5/Exemple/'

# Create the folder if it does not exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)
    
# Iterate through each row and download the file
for index, row in df_sample.iterrows():
    ftp_path = row['FTP_path']
    file_name = ftp_path.split('/')[-1]
    file_path = os.path.join(folder_path, file_name)
    urllib.request.urlretrieve(ftp_path, file_path)
```

# Sample ex

```python
# 定义vcf文件路径
vcf_file = '../Folder/File_5/Exemple/SRR21269155.annot.vcf.gz'

# 使用gzip库打开压缩的vcf文件并读取
with gzip.open(vcf_file, 'rt') as f:
    # 跳过以##开头的注释行
    lines = [line for line in f if not line.startswith('##')]
    # 获取以#开头的标题行
    header = lines[0].strip().split('\t')
    # 读取剩下的数据行
    data = [line.strip().split('\t') for line in lines[1:]]

# 将数据转换为pandas DataFrame
df = pd.DataFrame(data, columns=header)
df
```

```python
# 定义BAM文件路径
bam_file = '../Folder/File_5/Exemple/SRR21269155.bam'

# 打开BAM文件
bam = pysam.AlignmentFile(bam_file, 'rb')

# 从BAM文件中读取数据
data = []
for read in bam:
    data.append([read.query_name, read.reference_name, read.reference_start])

# 关闭BAM文件
bam.close()

# 将数据转换为pandas DataFrame
df = pd.DataFrame(data, columns=['read_name', 'ref_name', 'ref_start'])

df
```

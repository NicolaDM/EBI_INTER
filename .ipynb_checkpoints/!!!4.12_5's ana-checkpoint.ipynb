```python
import chardet

# 定义vcf文件路径
vcf_file = '../Folder/File_5/SRR21269155.bam'

# 使用gzip库打开压缩的vcf文件并读取
with gzip.open(vcf_file, 'rb') as f:
    # 检测文件编码
    file_content = f.read()
    result = chardet.detect(file_content)
    encoding = result['encoding']

    # 跳过以##开头的注释行
    lines = [line.decode(encoding) for line in file_content.splitlines() if not line.startswith(b'##')]
    # 获取以#开头的标题行
    header = lines[0].strip().split('\t')
    # 读取剩下的数据行
    data = [line.strip().split('\t') for line in lines[1:]]

# 将数据转换为pandas DataFrame
df = pd.DataFrame(data, columns=header)

# 打印DataFrame
df
```

####  Threshold:
- SB > 0, 
- coverage ratio < 0.1,
- AF > 0.001.

````python
import pandas as pd

threshold_AF = 0.001
threshold_SB = 0
threshold_COV = 0.1
````

## 1. Error position

````python
input_for_annot_err = "/nfs/research/goldman/zihao/Datas/p1_errorsProject_NEW/Folder_dataProcessing/ANNOT_dataProcessing_errorPos.txt"
input_for_cov_err = "/nfs/research/goldman/zihao/Datas/p1_errorsProject_NEW/Folder_dataProcessing/COV_dataProcessing_errorPos.txt"
````

````python
### For annot
df_err_annot = pd.read_csv(input_for_annot_err, sep='\t')

df_err_annot['ID_POS'] = df_err_annot['ID'].astype(str) + "_" + df_err_annot['Position'].astype(str)

df_err_annot['Flag_AF'] = df_err_annot['AF'] > threshold_AF
df_err_annot['Flag_SB'] = df_err_annot['SB'] > threshold_SB
df_err_annot.drop(['ID', 'Position','AF','SB'], axis=1, inplace=True)

print(df_err_annot)
print('=====================================')
### For coverage
df_err_cov = pd.read_csv(input_for_cov_err, sep='\t')

df_err_cov['ID_POS'] = df_err_cov['ID'].astype(str) + "_" + df_err_cov['Position'].astype(str)

df_err_cov['Flag_COV'] = df_err_cov['COV_Ratio'] < threshold_COV
df_err_cov.drop(['ID', 'Position', 'COV_Ratio'], axis=1, inplace=True)

print(df_err_cov)
print('=====================================')
### Merge
merged_df = pd.merge(df_err_annot, df_err_cov, on='ID_POS', how='inner')
merged_df.to_csv("/nfs/research/goldman/zihao/code/EBI_INTER/A_Datas/ERR_POS_all.txt", sep='\t', index=False)

merged_df
````

## 2. All position

````python
input_for_all = "/nfs/research/goldman/zihao/Datas/p1_errorsProject_NEW/Folder_dataProcessing/ANNOT_dataProcessing_allPos.txt"
````

````python
df_all = pd.read_csv(input_for_all, sep='\t')

df_all['ID_POS'] = df_all['ID'].astype(str) + "_" + df_all['Position'].astype(str)

df_all['Flag_AF'] = df_all['AF'] > threshold_AF
df_all['Flag_SB'] = df_all['SB'] > threshold_SB
df_all['Flag_COV'] = df_all['Cov_RATIO'] < threshold_COV
df_all.drop(['ID', 'Position','AF','SB','N','Cov_RATIO'], axis=1, inplace=True)

df_all.to_csv("/nfs/research/goldman/zihao/code/EBI_INTER/A_Datas/ALL_POS_all.txt", sep='\t', index=False)

df_all
````

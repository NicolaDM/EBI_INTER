# Visualization

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
from scipy.stats import gaussian_kde

hist_data = [df_err['error'].tolist()[0:1000], df_all['Ratio'].tolist()[0:1000]]
group_labels = ['Error position', 'All position']  # 数据集的名称
colors = ['#835AF1', '#B8F7D4']

fig = make_subplots(rows=2, cols=1, row_heights=[0.75, 0.25], shared_xaxes=True)

# 添加Density Plot子图
for data, label, color in zip(hist_data, group_labels, colors):
    kde = gaussian_kde(data)
    x_vals = np.linspace(min(data), max(data), 1000)
    y_vals = kde(x_vals)
    fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', fill='tozeroy', showlegend=False, line_color=color), 
                  row=1, col=1)

# 添加Box Plot子图
fig.add_trace(go.Box(x=df_err['error'].tolist()[0:1000], orientation='h', name=group_labels[0]), row=2, col=1)
fig.add_trace(go.Box(x=df_all['Ratio'].tolist()[0:1000], orientation='h', name=group_labels[1]), row=2, col=1)

fig.update_layout(height=600, title_text='Combined Plot')

fig.update_yaxes(range=[0, 2], row=1, col=1)
fig.update_xaxes(type='log', range=[np.log10(0.1), np.log10(np.max(np.concatenate(hist_data)))], row=1, col=1)
fig.update_xaxes(type='log', range=[np.log10(0.1), np.log10(np.max(np.concatenate(hist_data)))], row=2, col=1)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np

hist_data = [df_err['error'].tolist()[0:1000], df_all['Ratio'].tolist()[0:1000]]
group_labels = ['Error position', 'All position']  # 数据集的名称
colors = ['#835AF1', '#B8F7D4']

fig = make_subplots(rows=2, cols=1, row_heights=[0.75, 0.25], shared_xaxes=True)

# 添加Density Plot子图
for data, label, color in zip(hist_data, group_labels, colors):
    hist, edges = np.histogram(data, bins=30, density=True)
    fig.add_trace(go.Scatter(x=edges, y=hist, mode='lines', fill='tozeroy', showlegend=False, line_color=color), 
                  row=1, col=1)

# 添加Box Plot子图
fig.add_trace(go.Box(x=df_err['error'].tolist()[0:1000], orientation='h', name=group_labels[0]), row=2, col=1)
fig.add_trace(go.Box(x=df_all['Ratio'].tolist()[0:1000], orientation='h', name=group_labels[1]), row=2, col=1)

fig.update_layout(height=600, title_text='——————')

fig.update_xaxes(type='log', range=[np.log10(0.001), np.log10(np.max(edges))], row=1, col=1)
fig.update_xaxes(type='log', range=[np.log10(0.001), np.log10(np.max(edges))], row=2, col=1)

# 设置图形布局
fig.update_layout(
    title="Coverage density map over sample positions",
    yaxis=dict(
        title='Frequency',
        showline=True, showgrid=False,
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        )
    ),
    
    xaxis=dict(
        title='The ratio of coverage vs mean coverage',
        type='log',  # 将x轴设置为对数刻度
        showline=True, showgrid=False,
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        ),
    ),
    legend=dict(
        traceorder="normal",
        font=dict(size=12),
    ),

    plot_bgcolor='white',
    yaxis_gridcolor='lightgray', yaxis_gridwidth=0.5,
    xaxis_gridcolor='lightgray', xaxis_gridwidth=0.5,
)

fig.show()
```

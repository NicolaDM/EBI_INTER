```python
import plotly.graph_objects as go

fig = go.Figure()

for data, label, color in zip(hist_data, group_labels, colors):
    fig.add_trace(go.Histogram(x=data, histnorm='probability', name=label, marker_color=color, nbinsx=100))

fig.update_layout(barmode='overlay')
fig.write_image("output.png", format="png")
```

```python
import plotly.figure_factory as ff
import plotly.io as pio

hist_data = [(df_err[df_err['AF'] * 100 < 1]['AF'] * 100).tolist(), 
             (df_all[df_all['AF_Ratio'] * 100 < 1]['AF_Ratio'] * 100).tolist()]
group_labels = ['error position', 'all position'] # name of the dataset
colors = ['#7FA6EE', '#B8F7D4']

fig = ff.create_distplot(hist_data, group_labels, curve_type='normal', colors=colors, bin_size=.01)

# 设置图形布局
fig.update_layout(
    title="Allele Frequency",
    yaxis=dict(
        title='Probability density',
        showline=True, showgrid=False,
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        )
    ),
    xaxis=dict(
        title='Allele Frequency / %',
        # type='log',  # 对数刻度
        showline=True, showgrid=False,
        range=[0, 1],  # 对数刻度中间值，根据需求调整
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        )
    ),
    legend=dict(
        traceorder="normal",
        font=dict(size=12),
    ),
    plot_bgcolor='white',
    yaxis_gridcolor='lightgray', yaxis_gridwidth=0.5,
    xaxis_gridcolor='lightgray', xaxis_gridwidth=0.5,
)

# Save the plot as a PNG file
pio.write_image(fig, 'plot_for_AF_0-1.png')
```

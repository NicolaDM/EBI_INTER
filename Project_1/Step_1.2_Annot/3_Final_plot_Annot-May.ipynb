```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
from scipy.stats import gaussian_kde

hist_data = [(df_err[df_err['AF'] * 100 < 1]['AF'] * 100).tolist()[0:1000], 
             (df_all[df_all['AF_Ratio'] * 100 < 1]['AF_Ratio'] * 100).tolist()[0:1000]]
group_labels = ['error position', 'all position'] # name of the dataset
colors = ['#835AF1', '#B8F7D4']

fig = make_subplots(rows=2, cols=1, row_heights=[0.75, 0.25], shared_xaxes=True)

# 添加Density Plot子图
for data, label, color in zip(hist_data, group_labels, colors):
    kde = gaussian_kde(data)
    x_vals = np.linspace(min(data), max(data), 1000)
    y_vals = kde(x_vals)
    fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', fill='tozeroy', name=label, line_color=color), 
                  row=1, col=1)

# 添加Box Plot子图
fig.add_trace(go.Box(x=df_err['AF'].tolist()[0:1000], orientation='h', showlegend=False), row=2, col=1)
fig.add_trace(go.Box(x=df_all['AF_Ratio'].tolist()[0:1000], orientation='h', showlegend=False), row=2, col=1)

fig.update_layout(height=600, title_text='Combined Plot')

fig.update_yaxes(range=[0, 50], row=1, col=1)
fig.update_xaxes(type='log', range=[np.log10(0.005), 0.1], row=1, col=1)
fig.update_xaxes(type='log', range=[np.log10(0.005), 0.1], row=2, col=1)

fig.show()
```

```python
import plotly
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
from scipy.stats import gaussian_kde

hist_data = [(df_err[df_err['AF'] * 100 < 1]['AF'] * 100).tolist(), 
             (df_all[df_all['AF_Ratio'] * 100 < 1]['AF_Ratio'] * 100).tolist()]
group_labels = ['error position', 'all position'] # name of the dataset
colors = ['#835AF1', '#B8F7D4']

fig = make_subplots(rows=2, cols=1, row_heights=[0.75, 0.25], shared_xaxes=True)

# 添加Density Plot子图
for data, label, color in zip(hist_data, group_labels, colors):
    kde = gaussian_kde(data)
    x_vals = np.linspace(min(data), max(data), 1000)
    y_vals = kde(x_vals)
    fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', fill='tozeroy', name=label, line_color=color), 
                  row=1, col=1)

# 添加Box Plot子图
fig.add_trace(go.Box(x=df_err['AF'].tolist(), orientation='h', showlegend=False), row=2, col=1)
fig.add_trace(go.Box(x=df_all['AF_Ratio'].tolist(), orientation='h', showlegend=False), row=2, col=1)

fig.update_layout(height=600, title_text='Combined Plot')

fig.update_yaxes(range=[0, 50], row=1, col=1)
fig.update_xaxes(type='log', range=[np.log10(0.005), 0.1], row=1, col=1)
fig.update_xaxes(type='log', range=[np.log10(0.005), 0.1], row=2, col=1)

# 设置图形布局
fig.update_layout(
    title="Allele Frequency density map over sample positions",
    yaxis=dict(
        title='Probability density',
        showline=True, showgrid=False,
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        )
    ),
    xaxis=dict(
        title='Allele Frequency / %',
        # type='log',  # 对数刻度
        showline=True, showgrid=False,
        range=[0, 1],  # 对数刻度中间值，根据需求调整
        linewidth=2, linecolor='gray', ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='black',
        )
    ),
    legend=dict(
        traceorder="normal",
        font=dict(size=12),
    ),
    plot_bgcolor='white',
    yaxis_gridcolor='lightgray', yaxis_gridwidth=0.5,
    xaxis_gridcolor='lightgray', xaxis_gridwidth=0.5,
)

#fig.show()
#pio.write_image(fig, 'plot_for_AF_n.png')
plotly.offline.plot(fig, filename = "plot_for_AF_n.html")
```
